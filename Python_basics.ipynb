```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2 * np.pi, 0.2)
y = np.sin(x)
plt.plot(x, y)
plt.show()

```

```python
x = np.arange(0, 2 * np.pi, 0.2)
ysin = np.sin(x)
ycos = np.cos(x)

plt.plot(x, ysin)
plt.plot(x, ycos)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Sin & Cos')
plt.legend(['Sin','Cos'])
plt.show()
```

```python
fig, axs = plt.subplots(2,1)

plt.axes(axs[0])
plt.plot(x, ysin)
plt.title('Sin')
plt.axes(axs[1])
plt.plot(x, ycos)
plt.title('Cos')
plt.show()
```
